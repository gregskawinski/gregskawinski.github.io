```python
#ITERATORS and GENERATORS

my_list = [4, 7, 0, 1, 5]       
# create an iterator from the list
iterator = iter(my_list)

#get the first, second etc elements of the iterator
print(next(iterator))  # prints 4
print(next(iterator))  # prints 7
print(next(iterator))  # prints 0

# iterate through the elements of the iterator
for element in iterator: 
    print(element)

# custom iterator class,  __iter__() and the __next__() methods to be implemented

from itertools import count
# create an infinite iterator that starts at 1 and increments by 1 each time
infinite_iterator = count(1)

# print the first 5 elements of the infinite iterator
for i in range(5):
    print(next(infinite_iterator))

# GENERATORS
# generator is a function that returns an iterator that produces a sequence of values when iterated over.
# to produce a large sequence of values, but we don't want to store all of them in memory at once.

# Generators + the ‘yield’ Keyword
def simple_generator():
  yield 'apple'
  yield 'orange'
  yield 'pear'          #  yield statement insted of return statement

for fruit in simple_generator():        # Output apple orange pear
  print(fruit)

# generator 
def double_numbers(number):
    var = list(range(1,number+2))
    for i in var:
        yield i + i
# init the generator
generator = double_numbers(3)
# next element
print(next(generator))
print(next(generator))

# Even numbers Infinite Stream
def all_even():
    n = 0
    while True:
        yield n
        n += 2
# fractional range generator
def frange(start, stop, increment):
    x = start
    while x < stop:
        yield x
        x += increment
# Fibonacci generator
def fibonacci_numbers(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2

print(sum(square(fibonacci_numbers(10))))       #Pipelining Generators

# create the generator object
squares_generator = (i * i for i in range(5))
# iterate over the generator and print the values
for i in squares_generator:
    print(i)

# create generator object
values = (-x for x in [1,2,3,4,5])
gen_to_list = list(values)
print(gen_to_list)            # Output [-1, -2, -3, -4, -5]
```

```python
# CONTEXT MANAGER
# object that is used to manage resources, such as file handling, network connections, and database connections
# guarantees that the resources managed by the context manager are properly released at the end of the block, even if an exception occurs.
# if any error occurs the context manager mechanism automatically clean up the resources for you.

# Standard context manager for file operations
with open('hello.txt', 'w') as f:
    data = f.read()
    f.write('Hello Python!')

# Temporarily changing the working directory with a context manager and return to orginal after work done.
from contextlib import contextmanager

@contextmanager
def change_dir_contextmanager(destination):
    try:
        cwd = os.getcwd()
        os.chdir(destination)
        yield
    finally:
        os.chdir(cwd)

with change_dir_contextmanager('directory_1'):
    print(os.listdir())

# Network Operations 
import requests

@contextmanager
def network_request(url):
    response = requests.get(url)
    try:
        yield response
    finally:
        response.close()

# Usage
with network_request("https://example.com") as response:
    print(response.text)
    # Process the network response

# Sample db connection
from mysql_context_manager import MysqlConnector        # pip install mysql-context-manager

async with MysqlConnector(hostname="localhost") as conn:
    results = await conn.query("select username from users where is_bender = 1 order by username asc;")
    print(results[0]["username"])
```

```python
# Classes / OOP
# Private variables in python are not actually hidden fields like in other object oriented languages
# Using @property decorators to achieve getters and setters behaviour
# __new__ is responsible for creating a new instance of the class. It's a static method that takes the class itself (cls) as its first parameter.
# __init__ is called after the object has been created and is used to initialize its attributes.
# It's typically used for setting up instance-specific variables and performing other initialization tasks, takes self as the first parameter, 

class Person:
    
    # Class variable
    count = 0                               # Keeps track of the number of Person instances

    # constructor function
    def __init__(self, name):
        Person.count += 1                 # Increment the count of Person instances
        self.name = name                  # variable for name
        self.age = None                   # An example of an instance variable (age) with a default value
        self._internal = True             # designated for internal use, but doesn't stop you from accessing/modifileng
        ## private varibale or property in Python
        self.__datacamp = "Excellent"      # used to name mangling to avoid overide. python change the name
        
    # Getter function, to get the properties using an object
    @property
    def name(self):
        """Getter method for name."""
        return self._name
    # Setter function, to change the value  using an object
    @name.setter
    def name(self, value):
        """Setter method for name."""
        if not isinstance(value, str):
            raise TypeError('Expected a string')
        self._name = value
    # Deleter function
    @name.deleter
    def name(self):
        """Deleter method for name."""
        raise AttributeError("Can't delete attribute")
    
    # Instance methods
    def greet(self):
        """A simple greeting method."""
        return f"Hello, my name is {self._name}."

    @classmethod
    def get_instance_count(cls):
        """Class method to get the count of Person instances."""
        return cls.count
    
    # Magic methods, object’s Dunder
    def __str__(self):              # __str__: is whatever you think is that object in text form
        """String representation of the object."""
        return f"Person: {self._name}"
    
    def __repr__(self):            # __repr__: representation of python object usually eval will convert it back to that object
        """Formal string representation of the object."""
        return f"Person('{self._name}')"
    
    def __eq__(self, other):
        if isinstance(other, Person):
            return self._name == other.name
        return False
    
    def __len__(self):
        return len(self._name)
    
# User-defined exceptions
class CustomError(Exception):
    pass

# Class Instantiation
# Creating instances of the Person class
alice = Person("Alice")
bob = Person("Bob")
# Accessing the name property, Using the getter and setter for the 'name' property
bob.name                        # Get name, calls the getter
alice.name = "Alicia"           # Setting the name property
del bob.name                    # Deleting the name property
# Using the methods
print(alice.greet())            # Output: "Hello, my name is Alicia."
# Accessing the class variable
print(f"Number of Person instances: {Person.get_instance_count()}")

# Using dunder methods
print(alice)                # __str__
print(repr(bob))            # __repr__
print(alice == bob)         # __eq__
print(len(alice))           # Length of name, Using __len

```

```python
# DECORATORS
import time

# Basic Decorators
def uppercase_decorator(func):
   def wrapper(*args, **kwargs):
     result = func(*args, **kwargs)
     return result.upper()
   return wrapper

@uppercase_decorator
def greet(name):
    return f"Hello, {name}!"

greet("Romeo and Juliet")

# Junior-level decorator
def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} took {end_time - start_time} seconds to execute.")
        return result
    return wrapper
@measure_time
def junior_level_function():
    time.sleep(2)
junior_level_function()

# Intermediate-level decorator
def authenticate(func):
    def wrapper(*args, **kwargs):
        if is_user_authenticated():
            return func(*args, **kwargs)
        else:
            return "Access denied"
    return wrapper

@authenticate
def intermediate_level_function():
    return "Welcome to the secure area!"
def is_user_authenticated():
    # Check user authentication logic here
    return True
print(intermediate_level_function())

# Senior-level decorator for error handling
def handle_exceptions(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"An error occurred: {str(e)}")
    return wrapper

@handle_exceptions
def senior_level_function():
    result = 10 / 0  # Simulate a division by zero error
    return result
senior_level_function()

# Expert-level decorator for code profiling
import cProfile

def profile(func):
    def wrapper(*args, **kwargs):
        profiler = cProfile.Profile()
        profiler.enable()
        result = func(*args, **kwargs)
        profiler.disable()
        profiler.print_stats(sort="cumulative")
        return result
    return wrapper
@profile
def expert_level_function():
    # Your code here
    pass
expert_level_function()

import time
import logging
from functools import wraps

def timethis(func):
    '''
    Decorator that reports the execution time.
    '''
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(func.__name__, end-start)
        return result
    return wrapper

def logged(level, name=None, message=None):
    '''
    Add logging to a function. level is the logging
    level, name is the logger name, and message is the
    log message. If name and message aren't specified,
    they default to the function's module and name.
    '''
 
    def decorate(func):
        logname = name if name else func.__module__
        log = logging.getLogger(logname)
        logmsg = message if message else func.__name__
 
        @wraps(func)
        def wrapper(*args, **kwargs):
            log.log(level, logmsg)
            return func(*args, **kwargs)
        return wrapper
    return decorate

# Example use
@logged(logging.DEBUG)
def add(x, y):
    return x + y

@logged(logging.CRITICAL, 'example')
def spam():
    print('Spam!')
```
